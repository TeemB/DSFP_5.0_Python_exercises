```python


def is_prime(n):
    """Check if a number is prime."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# List to store consecutive prime pairs with gap >= 20
prime_gaps = []

previous_prime = 2  # Start from the first prime

# Loop through numbers up to 10,000
for num in range(3, 10001):
    if is_prime(num):
        gap = num - previous_prime
        if gap >= 20:
            prime_gaps.append((previous_prime, num))
        previous_prime = num  # Update previous prime

# Print some results
print(f"Found {len(prime_gaps)} prime pairs with gap ≥ 20.")

# Save to a text file
with open("prime_gaps.txt", "w") as f:
    for p1, p2 in prime_gaps:
        f.write(f"{p1}, {p2}\n")

print("Prime gaps saved to 'prime_gaps.txt'.")
```
