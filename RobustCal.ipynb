 1. Robust Calculator

```python
def add(a,b):
    """
    Converts both inputs to float and returns their sum.
    """
    a = float(a)          # Convert first operand to float
    b = float(b)          # Convert second operand to float
    return (a+b)          # Return the addition result


def subtract(a,b):
    """
    Converts both inputs to float and returns their difference (a - b).
    """
    a=float(a)            # Convert first operand to float
    b=float(b)            # Convert second operand to float
    return (a - b)        # Return the subtraction result


def multiply(a,b):
    """
    Converts both inputs to float and returns their product.
    """
    a = float(a)          # Convert first operand to float
    b = float(b)          # Convert second operand to float
    return (a * b)        # Return the multiplication result


def divide(a,b):
    """
    Attempts to divide a by b.
    Catches division-by-zero errors and prints an error message.
    """
    float(a)              # Convert a to float (result not reassigned)
    float(b)              # Convert b to float (result not reassigned)
    try:
        a/b               # Attempt division
    except ZeroDivisionError:
        print("Denominator cannot be zero")   # Handle division by zero
        return (a/b)      # Return the division result


def expression(exp):
    """
    Evaluates a mathematical expression provided as a string
    in the format: 'number operator number' (e.g., '3 + 4').
    """
    try:
        a, op,b = exp.split()  # Split expression into operands and operator
        a = float(a)          # Convert first operand to float
        b = float(b)          # Convert second operand to float
        
        if op == "+": 
            return add(a,b)   # Call addition function
        
        if op == "-": 
            return subtract(a,b)  # Call subtraction function
        
        if op == "*": 
            return multiply(a,b)  # Call multiplication function
        
        if op == "/": 
            return divide(a,b)    # Call division function
        
        return "invalid operator" # Handle unsupported operators
        
    except :
        return "Error invalid input"  # Handle invalid expression format


def manual_entry():
        """
        Prompts the user to manually enter two numbers and choose
        an operation from a menu.
        """
        try:
            a = float(input("input first number"))   # Get first number from user
            b = float(input("input second number"))  # Get second number from user
            
            # Display operation options
            print("Operation Menu")
            print("1 - Addition")
            print("2 - Subtraction")
            print("3 - Multiplication")
            print("4 - Division")
            
            selected_operator = input("Enter the number for the preferred operation")
            
            if  selected_operator == "1":
                return add(a,b)        # Perform addition
            if  selected_operator == "2":
                return subtract(a,b)   # Perform subtraction
            if  selected_operator == "3":
                return multiply(a,b)   # Perform multiplication
            if  selected_operator == "4":
                return divide(a,b)     # Perform division
            
            return "invalid choice"    # Handle invalid menu choice
        
        except:
            return "error: Invalid input"  # Handle input conversion errors
    

def calculate():
  """
  Main calculator loop.
  Allows the user to choose between manual entry and expression mode,
  performs calculations, and repeats until the user exits.
  """
  
  while True:
    mode = input("1. Manual entry  2.Expression" )  # Select calculator mode
        
    if mode == "2":
        exp = input("enter expression")   # Get expression input
        result = expression(exp)          # Evaluate expression
    elif mode == "1":
        result = manual_entry()           # Run manual entry mode
    else:
        result = "Invalid selection"      # Handle invalid mode
            
    print("Result:",result)               # Display result
    
    again = input("another calculation? y or n").lower()  # Ask to continue
    if again != "y":
        print("See you!!")                # Exit message
        break                             # Exit loop
        
calculate()  # Start the calculator


        

    
```
