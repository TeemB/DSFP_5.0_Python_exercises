Student Records

```python


student_record = []

#  Add Student
def add_student():
    """
    Adds a new student to the student_record list.
    Validates unique ID, positive age, and GPA within 0.0-4.0.
    """
    s_id = input("Enter student ID: ")

    # Check if ID already exists
    for s in student_record:
        if s["id"] == s_id:
            print("This ID already exists!")
            return

    s_name = input("Enter student name: ")

    # Safely get age
    try:
        s_age = int(input("Enter student age: "))
        if s_age <= 0:
            print("Age must be greater than zero")
            return
    except ValueError:
        print("Invalid input! Age must be a number.")
        return

    # Safely get GPA
    try:
        s_gpa = float(input("Enter GPA between 0.0 and 4.0: "))
        if s_gpa < 0.0 or s_gpa > 4.0:
            print("GPA must be between 0.0 and 4.0")
            return
    except ValueError:
        print("Invalid input! GPA must be a number.")
        return

    # Create and store the student
    student = {
        "id": s_id,
        "name": s_name,
        "age": s_age,
        "gpa": s_gpa
    }
    student_record.append(student)
    print("Student added successfully!")

# Update Student
def update_student():
    """
    Updates a student's name, age, or GPA by their ID.
    """
    s_id = input("Enter student ID to update: ")

    for s in student_record:
        if s["id"] == s_id:
            s_name = input("Enter new student name: ")

            # Safely get age
            try:
                s_age = int(input("Enter new student age: "))
                if s_age <= 0:
                    print("Age must be greater than zero")
                    return
            except ValueError:
                print("Invalid input! Age must be a number.")
                return

            # Safely get GPA
            try:
                s_gpa = float(input("Enter new GPA between 0.0 and 4.0: "))
                if s_gpa < 0.0 or s_gpa > 4.0:
                    print("GPA must be between 0.0 and 4.0")
                    return
            except ValueError:
                print("Invalid input! GPA must be a number.")
                return

            # Update student details
            s["name"] = s_name
            s["age"] = s_age
            s["gpa"] = s_gpa
            print("Student record updated successfully!")
            return

    print("Student not found!")

#View Student 
def view_student():
    """
    Views a student's information by their ID.
    """
    s_id = input("Enter student ID to view: ")

    for s in student_record:
        if s["id"] == s_id:
            print(s)
            return

    print("No student with such ID.")

#  Delete Student 
def delete_student():
    """
    Deletes a student from the record by their ID.
    """
    s_id = input("Enter student ID to delete: ")

    for s in student_record:
        if s["id"] == s_id:
            student_record.remove(s)
            print("Student record deleted!")
            return

    print("Student not found!")

#  List All Students 
def list_students():
    """
    Lists all students in the record.
    """
    if not student_record:
        print("No student records available.")
        return

    for s in student_record:
        print(s)

#Menu 
def student_record_management():
    """
    Main menu loop for managing students.
    """
    while True:
        print("\n===== Student Record Menu =====")
        print("1 - Add Student")
        print("2 - View Student")
        print("3 - Update Student")
        print("4 - Delete Student")
        print("5 - List All Students")
        print("6 - Exit")

        option = input("Choose a number from the menu: ")

        if option == "1":
            add_student()
        elif option == "2":
            view_student()
        elif option == "3":
            update_student()
        elif option == "4":
            delete_student()
        elif option == "5":
            list_students()
        elif option == "6":
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice! Please try again.")
#Run Program
student_record_management()
```
